```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/Prompt_engg_assgt final - Copy.xlsx')
```

```python
df.head()
```

```python
# Calculate the average for each row, considering columns starting with "Student_"
df['AVG'] = df.filter(regex='^Student_').mean(axis=1)

df.head()
```

```python
print(df.head())
```

```python
# Initialize or overwrite the 'DCS' column
df['DCS'] = 0.0  # Ensure the column is float for storing calculated values

# Calculate DCS for each row
for index, row in df.iterrows():
    # Extract the student name to match the respective column
    student_column = row['Student Name']  # E.g., 'Student_1', 'Student_2', etc.

    # Retrieve the value from the column corresponding to the student name
    if student_column in df.columns:
        matching_value = row[student_column]
        dcs = float(matching_value - row['AVG'])  # Calculate DCS
        df.at[index, 'DCS'] = dcs  # Assign DCS to the respective row
    else:
        # Handle cases where the student column is not found (optional)
        df.at[index, 'DCS'] = None

# Display the updated DataFrame (for testing/verification)
df.head()
```

```python
# Save the modified dataframe back to an Excel file
df.to_excel('output_with_updated_dcs.xlsx', index=False)
```
